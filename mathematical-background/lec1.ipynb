```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.fft import fft, fftfreq, ifft
```

```python
def sine_wave(t, frequency, phase):
    frequencies = t*frequency*2*np.pi
    y = np.sin(frequencies+phase)
    return y
```

```python
t = np.linspace(0, 2, 1024, endpoint=False)
y = sine_wave(t, 5, 0)

plt.plot(t, y)
plt.show()
```

```python
t = np.linspace(0, 2, 1024, endpoint=False)
y = sine_wave(t, 5, np.pi/4)

plt.plot(t, y)
plt.show()
```

```python
t = np.linspace(0, 1, 1024, endpoint=False)
y1 = sine_wave(t, 5, 0)
y2 = sine_wave(t, 5, np.pi/2)

plt.plot(t, y1)
plt.plot(t, y2)
plt.show()
```

```python
t = np.linspace(0, 1, 1024, endpoint=False)
y1 = sine_wave(t, 5, 0)
y2 = sine_wave(t, 5, np.pi)
y3 = y1+y2

plt.plot(t, y1)
plt.plot(t, y2)
plt.plot(t, y3)
plt.show()
```

```python
t = np.linspace(0, 2, 256, endpoint=False)
y1 = sine_wave(t, 5, 0)
sample_rate = 256/2
yf = fft(y1)
xf = fftfreq(256, 1/sample_rate)

plt.plot(xf, np.abs(yf))
plt.show()
```

```python
y2 = ifft(yf)

plt.plot(t, y2)
plt.show()
```
